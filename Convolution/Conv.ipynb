```python
import numpy as np
import matplotlib.pyplot as plt

fs = 10
t = np.arange(0, 50, 1 / fs)
N = len(t)

def x1(t):
    return 1 if t == 4 else 0

def x2(t):
    if t == 4:
        return 2
    if t == 10:
        return 1
    if t == 20:
        return 3
    return 0

def x3(t):
    if (t >= 4) and (t <= 4.5):
        return 2
    if t == 10:
        return 1
    if t == 20:
        return 3
    return 0

h = t ** 8.6 * np.exp(-t / 0.547)

def cnv(x, h):
    y = np.zeros(N)
    for n in range(N):
        for m in range(n):
            y[n] += h[m] * x(t[n - m])
    return y
```

```python
c1 = cnv(x1, h)
c2 = cnv(x2, h)
c3 = cnv(x3, h)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(t, c1, color = "red")
ax[0].set_title('conv1')
ax[1].plot(t, c2, color = "red")
ax[1].set_title('conv2')
ax[2].plot(t, c3, color = "red")
ax[2].set_title('conv3')

plt.show()
```
