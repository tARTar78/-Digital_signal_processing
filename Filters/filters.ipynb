```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("w.csv", header= None)
```

```python
x = np.array(df.loc[:,0])
x
```

```python
M = 50
N = len(df)
h1 = np.zeros(N)
f = 0.1
for i in range(M):
    h1[i] = np.sin(2*np.pi*f*(i-M/2))/((i-M/2)*np.pi)*(0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M))
h1[M//2] = 2*np.pi*f
h1 = h1/np.sum(np.abs(h1))
```

```python
M = 50
N = len(df)
h2 = np.zeros(N)
f = 0.4
for i in range(M):
    h2[i] = np.sin(2*np.pi*f*(i-M/2))/((i-M/2)*np.pi)*(0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M))
h2[M//2] = 2*np.pi*f
h2 = h2/np.sum(np.abs(h2))
for i in range(M):
    if(i % 2 == 0):
        h2[i] = -h2[i]
```

```python
M = 50
N = len(df)
h31 = np.zeros(N)
f = 0.3
for i in range(M):
    h31[i] = np.sin(2*np.pi*f*(i-M/2))/((i-M/2)*np.pi)*(0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M))
h31[M//2] = 2*np.pi*f
h31 = h31/np.sum(np.abs(h31))

M = 50
N = len(df)
h32 = np.zeros(N)
f = 0.4
for i in range(M):
    h32[i] = np.sin(2*np.pi*f*(i-M/2))/((i-M/2)*np.pi)*(0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M))
h32[M//2] = 2*np.pi*f
h32 = h32/np.sum(np.abs(h2))
for i in range(M):
    if(i % 2 == 0):
        h32[i] = -h32[i]
```

```python
def cnv(x, h):
    y = np.zeros(N)
    for n in range(N):
        for m in range(n):
            y[n] += x[m] * h[n - m] 
    return y
```

# Исходный сигнал

```python
import numpy as np
from pylab import *
from scipy import *
import os

def plotSpectrum(y,Fs):
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 frq = frq[range(n//2)] # one side frequency range

 Y = fft.fft(y)/n # fft computing and normalization
 Y = Y[range(n//2)]
 
 plot(frq,abs(Y),'r') # plotting the spectrum
 xlabel('Freq (Hz)')
 ylabel('|Y(freq)|')

Fs=1024.0
y = x
t = arange(0,len(y),1) # time vector

subplot(2,1,1)
plot(t,y)
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum(y,Fs)
show()
```
