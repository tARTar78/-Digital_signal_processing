```python
import numpy as np
import matplotlib.pyplot as plt

fs = 400 
t = np.arange(0, 1, 1/fs)  
signal = np.cos(2*np.pi*1*t) + np.cos(2*np.pi*40*t) 

freqs = np.fft.fftfreq(len(signal), 1/fs)
fft_signal = np.fft.fft(signal)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Сигнал')
plt.xlabel('Время, с')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(freqs[:len(freqs)//2], 2*np.abs(fft_signal)[:len(freqs)//2]/fs)
plt.title('Спектр сигнала')
plt.xlabel('Частота, Гц')
plt.grid()

plt.tight_layout()
plt.show()
```
